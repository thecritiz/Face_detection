```python
import cv2
import os

def get_next_user_id(save_path):
    count_file = os.path.join(save_path, "last_user_id.txt")
    
    # Read the last used user ID from file
    if os.path.exists(count_file):
        with open(count_file, 'r') as f:
            last_user_id = int(f.read().strip())
    else:
        last_user_id = 0  # Start with user ID 1 if the file does not exist
    
    # Increment to the next user ID
    next_user_id = last_user_id + 1
    
    # Save the next user ID back to the file
    with open(count_file, 'w') as f:
        f.write(str(next_user_id))
    
    return next_user_id

def generate_dataset(camera_index=0, num_samples=200, save_path="D:/Projects/Recognition_System/data/"):
    # Check if the save directory exists, and create it if it doesn't
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    
    def face_cropped(img):
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_classifier.detectMultiScale(gray, 1.3, 5)
        # scaling factor = 1.3, minimum neighbors = 5
         
        if len(faces) == 0:
            return None
        for (x, y, w, h) in faces:
            cropped_face = img[y:y+h, x:x+w]
            return cropped_face
    
    user_id = get_next_user_id(save_path)
    cap = cv2.VideoCapture(camera_index)
    img_id = 0
    
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image. Exiting...")
            break
        
        cropped_face = face_cropped(frame)
        if cropped_face is not None:
            img_id += 1
            face = cv2.resize(cropped_face, (200, 200))
            face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
            file_name_path = os.path.join(save_path, f"user.{user_id}.{img_id}.jpg")
            cv2.imwrite(file_name_path, face)
            cv2.putText(face, str(img_id), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
             
            cv2.imshow("Cropped Face", face)
             
        if img_id >= num_samples:
            break
        
        if cv2.waitKey(1) == 13:  # Enter key to exit
            break
    
    cap.release()
    cv2.destroyAllWindows()
    print(f"Collected {img_id} samples for user {user_id}. Process completed.")

generate_dataset()
```

```python
import os
import cv2
import numpy as np
from PIL import Image

def train_classifier(data_dir):
    faces = []
    ids = []
    
    for root, dirs, files in os.walk(data_dir):
        for file in files:
            if file.endswith('.jpg'):  # Ensure only .jpg files are processed
                image_path = os.path.join(root, file)
                try:
                    img = Image.open(image_path).convert('L')
                except IOError as e:
                    print(f"Error opening image {image_path}: {e}")
                    continue
                
                imageNp = np.array(img, 'uint8')
                id = int(os.path.split(image_path)[1].split(".")[1])
                
                faces.append(imageNp)
                ids.append(id)
    
    if len(faces) == 0:
        print("No images found to train.")
        return

    clf = cv2.face.LBPHFaceRecognizer_create()
    clf.train(faces, np.array(ids))
    clf.write("classifier.xml")
    print("Training complete and classifier saved.")

# Example usage
train_classifier("D:/Projects/Recognition_System/data")
```

```python
import cv2

def draw_boundary(img, classifier, scaleFactor, minNeighbors, color, text, clf):
    if img is None or img.size == 0:
        print("Error: Empty image frame.")
        return img  # Skip processing if the image is empty
    
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    features = classifier.detectMultiScale(gray_img, scaleFactor, minNeighbors)
    
    for (x, y, w, h) in features:
        cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
        
        id, pred = clf.predict(gray_img[y:y + h, x:x + w])
        confidence = int(100 * (1 - pred / 300))
        
        if confidence > 80:
            if id == 1:
                cv2.putText(img, "Aditya", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
            elif id == 2:
                cv2.putText(img, "Amoolya", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
            elif id == 3:
                cv2.putText(img, "Anshul", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
            elif id == 4:
                cv2.putText(img, "Sumith", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
                elif id == 5:
                cv2.putText(img, "hirlyn", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 1, cv2.LINE_AA)
        else:
            cv2.putText(img, "UNKNOWN", (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 1, cv2.LINE_AA)
    
    return img

# Load face cascade
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
if faceCascade.empty():
    print("Error: Could not load face cascade file!")
    exit()

# Load LBPH face recognizer
clf = cv2.face.LBPHFaceRecognizer_create()
try:
    clf.read("classifier.xml")
except Exception as e:
    print(f"Error loading classifier: {e}")
    exit()

# Start video capture
video_capture = cv2.VideoCapture(0)  # Change index if necessary
if not video_capture.isOpened():
    print("Error: Could not access the camera.")
    exit()

while True:
    ret, img = video_capture.read()
    if not ret or img is None or img.size == 0:
        print("Error: Failed to capture image.")
        break
    
    img = draw_boundary(img, faceCascade, 1.3, 6, (255, 255, 255), "Face", clf)
    cv2.imshow("Face Detection", img)
    
    if cv2.waitKey(1) == 13:  # Break on Enter key
        break

# Release video capture and close windows
video_capture.release()
cv2.destroyAllWindows()
```
